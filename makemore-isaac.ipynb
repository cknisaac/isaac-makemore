```python
b = {}
for w in words:
    chs= ["<S>"] + list(w) + ["<E>"] # add special start and end characters (list plus list = list)
    for ch1, ch2 in zip(chs, chs[1:]): #e.g. Emma -> Emma , mma: zip auto stops iterating when the shorter list runs out of elements
        bigram = (ch1, ch2) 
        b[bigram] = b.get(bigram, 0) + 1 # this creates a counter (everytime bigram found in dataset, increase by 1 and return 0 default if bigram not found)) output: {tuple:int}
       
```

```python
import torch
```

```python
N = torch.zeros((28,28), dtype=torch.int32) # create a 28x28 array that maps bigrams (column: 1st word, row: 2nd word)
```

```python
chars = sorted(list(set("".join(words)))) 
string_to_index = {s:i for i,s in enumerate(chars)}
string_to_index["<S>"] = 26
string_to_index["<E>"] = 27
```

```python
string_to_index
```

```python
for w in words:
    chs= ["<S>"] + list(w) + ["<E>"] # add special start and end characters (list plus list = list)
    for ch1, ch2 in zip(chs, chs[1:]): #e.g. Emma -> Emma , mma: zip auto stops iterating when the shorter list runs out of elements
        idx1 = string_to_index[ch1]
        idx2 = string_to_index[ch2]
        N[idx1, idx2] += 1 # this creates a counter (everytime bigram found in dataset, increase by 1 and return 0 default if bigram not found)) output: {tuple:int}
       
 # this creates a counter (everytime bigram found in dataset, increase by 1 and return 0 default if bigram not found)) output: {tuple:int}
       
```
